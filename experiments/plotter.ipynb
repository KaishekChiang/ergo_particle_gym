```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 22})
```

## simple_graph_large (sgl) analysis

```python
heuristic_sgl_file = "expdata.2018-10-25.heuristic_simple_graph_large.0/heuristic_simple_graph_large_50_episode_batch_reward_stats.pkl"
mcgroup_sgl_file = "expdata.2018-10-25.mcgroup_simple_graph_large.0/mcgroup_simple_graph_large_50_episode_batch_reward_stats.pkl"
maddpg_sgl_file = "expdata.2018-10-25.maddpg_simple_graph_large.0/maddpg_simple_graph_large_50_episode_batch_reward_stats.pkl"

# load reward
heuristic_sgl_batch_rewards = pickle.load(open(heuristic_sgl_file, "rb"))
mcgroup_sgl_batch_rewards = pickle.load(open(mcgroup_sgl_file, "rb"))
maddpg_sgl_batch_rewards = pickle.load(open(maddpg_sgl_file, "rb"))
```

```python
# plot
sgl_max_rewards = 1000
sgl_n_points = len(heuristic_sgl_batch_rewards)
plt.figure(figsize=(20,12))
plt.plot(
    [i for i in range(len(heuristic_sgl_batch_rewards))], [rew[0] for rew in heuristic_sgl_batch_rewards], '.-',
    [i for i in range(len(mcgroup_sgl_batch_rewards))], [rew[0] for rew in mcgroup_sgl_batch_rewards],'x-',
    [i for i in range(len(maddpg_sgl_batch_rewards))], [rew[0] for rew in maddpg_sgl_batch_rewards],'v-', 
    [i for i in range(sgl_n_points)], [sgl_max_rewards for i in range(sgl_n_points)],markersize=8)
plt.xlabel("Episodes in 50 Episode Increments")
plt.ylabel("Mean Episode Reward")
plt.title("Learning Curves for Scenario simple_graph_large")
plt.legend(["heuristic", "x-entropy", "deep-rl", "maximum"])
plt.show()
```
